```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
base_url = "https://news.abplive.com/news/world/feed"

response = requests.get(base_url)

content = response.content

parsed_content = bs(content, 'html.parser')

item_content = parsed_content.find_all("item")
```

```python
list_news = []

for item in item_content:
  title = item.find_all("title")[0].get_text().strip()
  link = item.find_all("guid")[0].get_text().strip()
  pubilsh = item.find_all("pubdate")[0].get_text().strip()
  description = item.find_all("description")[0].get_text().strip()
  img = item.find_all("media:thumbnail")
  if img:
    image = img[0].get("url")
  else:
    image = ""
  temp = {
    "title": title,
    "link": link,
    "pubilsh": pubilsh,
    "description": description,
    "image": image
  }
  
  list_news.append(temp)
```

```python
df = pd.DataFrame(list_news)
df
```

```python
df.to_csv("data/abpnews_temp.csv", sep="\t", index=False)
```
