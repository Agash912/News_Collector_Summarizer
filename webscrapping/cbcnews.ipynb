```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
base_url = "https://www.cbc.ca/webfeed/rss/rss-sports"

response = requests.get(base_url)

content = response.content

parsed_content = bs(content, 'html.parser')

item_content = parsed_content.find_all("item")
```

```python
list_news = []

for item in item_content:
  title = item.find_all("title")[0].get_text().strip()
  link = item.find_all("link")[0].get_text().strip()
  pubilsh = item.find_all("pubdate")[0].get_text().strip()
  tag = bs(item.find_all("description")[0].get_text())
  description = tag.find("p").get_text()
  image = tag.find("img").get("src")

  temp = {
    "title": title,
    "link": link,
    "pubilsh": pubilsh,
    "description": description,
    "image": image
  }
  
  list_news.append(temp)
```

```python
df = pd.DataFrame(list_news)
df
```

```python
df.to_csv("data/cbcnews_temp.csv", sep="\t", index=False)
```
