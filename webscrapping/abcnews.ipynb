```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
base_url = "https://abcnews.go.com/abcnews/topstories"

response = requests.get(base_url)

content = response.content

parsed_content = bs(content, 'html.parser')

item_content = parsed_content.find_all("item")
```

```python
list_news = []

for item in item_content:

  title = item.find_all("title")[0].get_text().strip()
  link = item.find_all("guid")[0].get_text().strip()
  pubilsh = item.find_all("pubdate")[0].get_text().strip()
  description = item.find_all("description")[0].get_text().strip()
  image = item.find_all("media:thumbnail")[0].get("url").strip()
  
  temp = {
    "title": title,
    "link": link,
    "pubilsh": pubilsh,
    "description": description,
    "image": image
  }
  
  list_news.append(temp)
```

```python
list_news[0]
```

```python
df = pd.DataFrame(list_news)
df
```

```python
df.to_csv("data/abcnews_temp.csv", sep="\t", index=False)
```

```python
detailed_news = []
for news in list_news:
  link = news["link"]
  # print(link)
  response_news = requests.get(link)

  content_news = response_news.content

  parsed_content_news = bs(content_news, 'html.parser')

  h1 = parsed_content_news.find_all("h1")
  if h1 == []:
    continue
  title = h1[0].get_text()
  
  content_tag = parsed_content_news.find_all("div", {"data-testid":"prism-article-body"})
  paras = content_tag[0].find_all("p")
  print(paras)
  disc = ""
  for p in paras:
    disc += str(p.get_text())
    
  detailed_news.append({"title": title, "disc": disc})
```

```python
df_disc = pd.DataFrame(detailed_news)
df_disc
```
